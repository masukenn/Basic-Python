```python
df_02 = pd.read_csv("C:\\Users\\kento\\Basic-Python\\colab\\quizzes\\matplotlib_quiz\\hist_data.csv")
plt.hist(df_02['aデータ'],bins = 30,alpha = 0.5,label='aデータ')
plt.hist(df_02['bデータ'],bins = 30,alpha = 0.5,label='bデータ')
plt.title('aデータとbデータの温度に関する度数分布')
plt.xlabel('温度℃')
plt.ylabel('度数')
plt.legend()


```

```python
df_03 = pd.read_csv("C:\\Users\\kento\\Basic-Python\\colab\quizzes\\matplotlib_quiz\\scatter_data.csv")
x = df_03['length (cm)']
y = df_03['width (cm)']
color_values = df_03['class']
plt.xlabel('length(cm)')
plt.ylabel('width(cm)')
plt.title('クラス0～2のlengthとwidthに関する散布図')
plt.grid()
plt.scatter(x,y, c=color_values)
```

```python
df_01 = pd.read_csv("C:\\Users\\kento\\Basic-Python\\colab\\quizzes\\matplotlib_quiz\\line_data.csv")
x = df_01['期間']
y = df_01['件数']
z = df_01['費用']
plt.plot(x, y, marker = 'o')
plt.xlabel('期間')
plt.xticks(rotation=45)
plt.ylabel('件数')
plt.grid(True)
plt.title('期間ごとの件数の推移')
plt.show()


```
